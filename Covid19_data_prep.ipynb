```
import scipy.stats as scpst
```

```
#### not the best name for it. just truncates the time series by initiating t=0 at 
#### state_case
def per_capita_arr(m, m2, start_case): 
  case_where = np.where(m  >= start_case) 
  
  new_m2 = m2[case_where]
  return new_m2
```

```
x_num_cases = 100 
#wid_ts = 30

corre_coeffs_arr = np.zeros(st_abvs_us_dat.size) 


for i in range(0,st_abvs_us_dat.size):
   ####start the clock after the nth case 
  st_where = np.where(USA_ST_C19_ARR[USA_ST_C19_ARR['state'] == st_abvs_us_dat[i]]['positive'].values  >= x_num_cases) 
  sk_where = np.where(c19_skorea['confirmed'].values  >= x_num_cases) 
  
  ###get the values that meet the criteria 
  sk = sk_where[0][0]
  us_in =st_where[0][0] 

  wid_ts = st_where[0][-1] - st_where[0][0] 

  ####us array
  usst_ts = USA_ST_C19_ARR[USA_ST_C19_ARR['state'] == st_abvs_us_dat[i]]['pos_per_cap'].values
  usst_ts_in = usst_ts[st_where] 

  ####sk array
  sk_ts = c19_skorea['confirmed_per_cap'].values
  sk_ts_in = sk_ts[sk_where] 


  ###standardizing 
  usst_stand = usst_ts_in[0:wid_ts ] - np.mean(usst_ts_in[0:wid_ts ]))/(np.std(usst_ts_in[0:wid_ts ]))
  sk_stand = sk_ts_in[0:wid_ts ] - np.mean(sk_ts_in[0:wid_ts ]))/(np.std(sk_ts_in[0:wid_ts ])) 


   ####calculating pearson coefficient 
  corre_coeffs_arr[i] = np.corrcoef(usst_stand, sk_stand )[0,1]
```
